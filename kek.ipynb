```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Создаем данные для графика
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Создаем график с помощью Seaborn
sns.lineplot(x=x, y=y)

ax = plt.gca()

ax.set_facecolor('black')

for i in range(1, 11):
    alpha = i / 10.0 
    ax.axhline(y=i, color='pink', alpha=alpha, linestyle='--')
    ax.axvline(x=i, color='white', alpha=alpha, linestyle='--')

plt.show()
```

```python
df = sns.load_dataset('titanic')
df
```

```python
import pandas as pd
age_bins = [0, 18, 35, 60, 100]
age_labels = ['Дети', "Молодые", " Взрослые", "Пожилые"]
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
df
```

```python
#1

kek = sns.color_palette('tab10')
sns.set_palette(palette=kek)

sns.set_color_codes('dark')
sns.set_style('whitegrid')


ax = sns.barplot(x=df.groupby('age_group')['survived'].count().index, y=df.groupby('age_group')['survived'].count().values, 
                 hue=df.groupby('age_group')['survived'].count().index, legend="brief")
ax.set_title('Процент выживших в зависимости от возрастной группы')
ax.legend(title='Возрастные группы')
# ax.bar_label(ax.containers[0], fontsize=10)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), 
                textcoords='offset points') #це чат гпт так как не выводит цифры я не знаю чо делац
plt.figure(figsize=(20,20))
```
